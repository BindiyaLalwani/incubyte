```python
class Space:
    def __init__(self, x, y, z, dirr):
        self.x = x
        self.y = y
        self.z = z
        self.dirr = dirr

    def MF(self):
        if self.dirr == 'n':
            self.y += 1
        elif self.dirr == 's':
            self.y -= 1
        elif self.dirr == 'e':
            self.x += 1
        elif self.dirr == 'w':
            self.x -= 1
        elif self.dirr == 'u':
            self.z += 1
        elif self.dirr == 'd':
            self.z -= 1

    def MB(self):
        if self.dirr == 'n':
            self.y -= 1
        elif self.dirr == 's':
            self.y += 1
        elif self.dirr == 'e':
            self.x -= 1
        elif self.dirr == 'w':
            self.x += 1
        elif self.dirr == 'u':
            self.z -= 1
        elif self.dirr == 'd':
            self.z += 1

    def TL(self):
        if self.dirr == 'n':
            self.dirr = 'w'
        elif self.dirr == 's':
            self.dirr = 'e'
        elif self.dirr == 'e':
            self.dirr = 'n'
        elif self.dirr == 'w':
            self.dirr = 's'
        elif self.dirr == 'u':
            self.dirr = 'w'
        elif self.dirr == 'd':
            self.dirr = 'e'

    def TR(self):
        if self.dirr == 'n':
            self.dirr = 'e'
        elif self.dirr == 's':
            self.dirr = 'w'
        elif self.dirr == 'e':
            self.dirr = 's'
        elif self.dirr == 'w':
            self.dirr = 'n'
        elif self.dirr == 'u':
            self.dirr = 'e'
        elif self.dirr == 'd':
            self.dirr = 'w'

    def TU(self):
        if self.dirr not in ['d']:
            self.dirr = 'u'

    def TD(self):
        if self.dirr not in ['u']:
            self.dirr = 'u'

    def coo(self, co):
        if co == 'f':
            self.MF()
        elif co == 'b':
            self.MB()
        elif co == 'l':
            self.TL()
        elif co == 'r':
            self.TR()
        elif co == 'u':
            self.TU()
        elif co == 'd':
            self.TD()
```

```python
def coo(x, y, z, dirr, co):
    bind = Space(x, y, z, dirr)

    for cm in co:
        bind.coo(cm)

    return bind.x, bind.y, bind.z, bind.dirr
```

```python
x = 0
y = 0
z = 0
d = 'n'
co = []
n=int(input("Enter number of commands:"))
for i in range(n):
    v=input("Enter direction variable using above commands:")
    if v not in ['d','l','u','f','b','r']:
        print("Wrong Direction")
    else:
        co.append(v)

xx, yy, zz, dd = coo(x, y, z, d, co)

print("Final position:",xx, yy, zz)
print("Final direction:", dd)
```
